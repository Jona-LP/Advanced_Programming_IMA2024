## _Gráfica de una Elipse_

Para este ejercicio se considera la elipse de ecuación
$$
\dfrac{x^{2}}{9} + \dfrac{y^{2}}{25} = 1.
$$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (8, 4)

# Parámetros de la elipse.
a = 3
b = 5

# Puntos en el eje X.
x = np.linspace(-a, a, 100)

# Calcula los valores en Y correspondientes a cada punto del dominio.
y = np.sqrt(b**2 * (1 - (x**2/a**2)))

plt.plot(x, y, 'g', label=r"$\dfrac{x^{2}}{9} + \dfrac{y^{2}}{25}=1$")
plt.plot(x, -y, 'g')
plt.axis('equal')
plt.title('«Gráfica Elipse»')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.grid()
plt.show()
```

## _Gráfica de una función en $\mathbb{R}^{3}$_

Para este ejercicio de considera la función
$$
f(x,y) = 9x^{2} - 5y^{2}.
$$

```python
import numpy as np
import matplotlib.pyplot as plt


# Genera los ejes en 3D.
ax = plt.axes(projection='3d')

# Genera los datos para los puntos f(x,y)=z.
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
z = 9*(x**2) - 5*(y**2)

ax.plot(x, y, z, 'b', label=r"$f(x,y) = 9x^{2} - 5y^{2}$")
ax.set_title('«Gráfica Función 3D»')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
plt.legend()
plt.show()
```

# **Homework \#2**✏️


```python
import numpy as np

class Point2D:
  def __init__(self, x=0.0, y=0.0):
    self.x = x
    self.y = y

  def __str__(self):
    return f'({self.x}, {self.y})'

  def distance(self, puntofinal):
    """
    - Computa la distancia entre self y el punto final.
    - Toma como input otro objeto de la clase Punto2D.
    - Entrega la distancia como float.
    """
    return np.sqrt((self.x - puntofinal.x)**2 + (self.y - puntofinal.y)**2)
```

```python
A = Point2D(1, 1)
B = Point2D(1, 3)
print(A.distance(B))
```

### __Ejercicio 3.1__

```python
import numpy as np

class Circulo:
  def __init__(self, radio):
    self.radio = radio

  def area(self):
    return np.pi * self.radio**2

  def perimetro(self):
    return 2 * np.pi * self.radio
```

```python
# Test
C = Circulo(3)
print(C.area())
print(C.perimetro())
```

### __Ejercicio 3.2__

```python
import numpy as np

class Punto3D:
  def __init__(self, x=0.0, y=0.0, z=0.0):
    self.x = x
    self.y = y
    self.z = z

  def distancia(self, puntofinal):
    """
    - Computa la distancia entre self y el punto final.
    - Toma como input otro objeto de la clase Punto3D.
    - Entrega la distancia como float.
    """
    return np.sqrt((self.x - puntofinal.x)**2 + (self.y - puntofinal.y)**2 + (self.z - puntofinal.z)**2)
```

```python
# Test
A = Punto3D(1, 2, 0)
B = Punto3D(3, 0, 0)

print(A.x, A.y, A.z)
print(A.distancia(B))
```

### __Ejercicio 3.3__

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (8, 4)

# Puntos en el eje X.
x = np.linspace(-10, 10, 100)

# Calcula los valores en Y correspondientes a cada punto del dominio.
y_1 = x**2 + 2*x + 3 # Fuciȯn f(x)
y_2 = 1 - (x**2)     # Funciȯn g(x)

plt.plot(x, y_1, 'r--', label=r"$f(x) = x^{2} + 2x + 3$")
plt.plot(x, y_2, 'g', label=r"$g(x) = -x^{2} + 1$")

plt.title('«Gráficos de Funciones $f(x)$ y $g(x)$»')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.grid()
plt.show()
```
